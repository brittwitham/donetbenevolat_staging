```python
import pandas as pd
```

```python
# REmoves any "unnamed" columns

def remove_unnamed_columns(df):
    return df.loc[:,~df.columns.str.startswith('Unnamed:')]
```

```python
test_df_path = "/Users/brittany/repos/donetbenevolat/tables/2018-AvgAmtBarriers_FR.csv"
test_df = pd.read_csv(test_df_path).pipe(remove_unnamed_columns)
```

```python
test_translations = {
    'Total donation amount': "montant total du don",
    'Average Amount Donated by Donors Reporting Barriers': "Montant moyen donné par les donateurs faisant état d'obstacles"
}
```

```python
gender_translations = {
    'Male': "Homme",
    'Female': "Female"
}
```

```python
def translate(df, translations):
    translated_df = df.replace(translations)
    return translated_df
```

```python
def package_translations(*lists):
    dict = {}
    for d in lists:
        dict.update(d)

    return dict
```

```python
combined_translations = package_translations(gender_translations, test_translations)
combined_translations
```
